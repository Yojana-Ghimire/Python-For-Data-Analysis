```python
data1= {"Emp Id":["E01","E02","E03","E04","E05","E06"],
       "Names":["Ram","Shyam","Rahul","Vishal","Ravi","John"]}
data2 = {"Emp Id": ["E01","E07","E03","E04","E08","E06"],"Salary":[45000,50000,60000,70000,80000,90000]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
print(df1)
print()
print(df2)
```

```python
#merge
print(pd.merge(df1,df2,on ="Emp Id",how = "left"))
```

```python
print(pd.merge(left = df1, right=df2, on="Emp Id",how="right"))
```

```python
sales = {"Fruits":["Mango","Apple","Banana","litchi","Grapes"],
        "Price":[100,150,112,200,250],
        "Quantity":[10,15,5,20,10]}
df= pd.DataFrame(sales)
df2=df.copy()
```

```python
dict = {"Keys":["k1","k2","k2","k1","k3"],
       "Names":["John","Ben","David","Peter","Josh"],
       "Houses":["Red","Blue","Green","Red","Black"],
       "Grades":["3rd","8th","2nd","3rd","2nd"]}
df=pd.DataFrame(dict)
print(df)
```

```python
#pivot tables:change rows into columns  
print(df.pivot(index="Keys",columns="Names",values=["Houses","Grades"]))
```

```python
dict = {
       "Names":["John","Ben","David","Peter","Josh"],
       "Houses":["Red","Blue","Green","Red","Black"],
       "Grades":["3rd","8th","2nd","3rd","2nd"]}
df=pd.DataFrame(dict)
print(df)
```

```python
#Melt: melting into 2 columns
print(pd.melt(df,id_vars=["Names"],value_vars="Houses&Grades",value_name="Values"))
```
